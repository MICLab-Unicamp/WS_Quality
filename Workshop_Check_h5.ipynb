## Read and visualize h5 data

```python
import h5py 
```

```python
#import nibabel as nib
import numpy as np
import os
import sys
import matplotlib.pyplot as plt
```

```python
DATASET_STRUCTURED_DIR = '/home/ubuntu/volume1/BrainHack/data/BrainHack_quality_'
```

```python
def read_h5_data(group,batch_size):
    
    # address to where you want to save the hdf5 file
    hdf5_path = DATASET_STRUCTURED_DIR+group+".hdf5"
    # open a hdf5 file and create datasets
    with h5py.File(hdf5_path, 'r') as hdf5_file:
        print(hdf5_file["labels"].shape)
        
        for ns in np.arange(0,hdf5_file["labels"].shape[0],batch_size):
            x = hdf5_file["img"][ns:ns+batch_size,:,:,:]
            xmin = np.min(x)
            xmax = np.max(x)
            if xmax-xmin == 0:
                print('img',xmin,xmax)


    return
```

```python
def visualize_h5_data(group,batch_size):
 
    fig=plt.figure(figsize=(15, 20))
    columns = 5; rows = 8

    # address to where you want to save the hdf5 file
    hdf5_path = DATASET_STRUCTURED_DIR+group+".hdf5"
    # open a hdf5 file and create datasets
    with h5py.File(hdf5_path, 'r') as hdf5_file:
        print(hdf5_file["labels"].shape)
        
        for ns in np.arange(batch_size,hdf5_file["labels"].shape[0],batch_size):
            fig.add_subplot(rows, columns, (ns//batch_size))
            x = hdf5_file["img"][ns,:,:,:]
            plt.imshow(x[:,:,0], cmap='gray')
            
    fig.add_subplot
    return
```

```python
visualize_h5_data('train',30800//40)
```
